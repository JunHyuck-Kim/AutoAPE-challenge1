```python
import tensorflow as tf
from keras.preprocessing.image import load_img, array_to_img, img_to_array
from keras.layers import Conv2D, MaxPooling2D, UpSampling2D, Input
import glob
import cv2
from keras.models import Sequential, Model
from scipy.ndimage import rotate
```

```python
train=glob.glob('/kaggle/working/denoising-dirty-documents/train/*.png')
clean=glob.glob('/kaggle/working/denoising-dirty-documents/train_cleaned/*.png')
test=glob.glob('/kaggle/working/denoising-dirty-documents/test/*.png')
```

```python
X = [] 
X_target = [] 

for img in train:
    img = load_img(img, grayscale=True,target_size=(420,540))
    img = img_to_array(img).astype('float32')/255.
    
    X.append(img)
    
    
for img in clean:
    img = load_img(img, grayscale=True,target_size=(420,540))
    img = img_to_array(img).astype('float32')/255.
    
    X_target.append(img)
    
X = np.array(X) 
X_target = np.array(X_target)
```

```python
def build_autoenocder():
    input_img = Input(shape=(420,540,1), name='image_input')
    
    #enoder 
    x = Conv2D(32, (3,3), activation='relu', padding='same', name='Conv1')(input_img)
    x = Conv2D(64, (3,3), activation='relu', padding='same', name='Conv2')(x)
    x = MaxPooling2D((2,2), padding='same', name='pool1')(x)
    
    #decoder
    x = Conv2D(64, (3,3), activation='relu', padding='same', name='Conv3')(x)
    x = Conv2D(32, (3,3), activation='relu', padding='same', name='Conv4')(x)
    x = UpSampling2D((2,2), name='upsample2')(x)
    x = Conv2D(1, (3,3), activation='sigmoid', padding='same', name='Conv5')(x)
    
    #model
    autoencoder = Model(inputs=input_img, outputs=x)
    return autoencoder

model = build_autoenocder()
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
history = model.fit(X, X_target, epochs=300, batch_size=2)
```

```python
import re

num=[]
predicted_list = [] 
for img in test:
    num.append(re.findall("\d+",img))
    img = load_img(img, color_mode='grayscale')
    img = img_to_array(img).astype('float32') / 255.0
    predicted_img = model.predict(np.expand_dims(img, axis=0))
    predicted_list.append(predicted_img)
```

```python
num=sum(num,[])
print(num)
```

```python
output=[['id', 'value']]
```

```python
i=0
for img in predicted_list:
    for j in range(len(img[0][0])):
        for k in range(len(img[0])):
            output.append([str(num[i])+'_'+str(k+1)+'_'+str(j+1), img[0][k][j].item()])
    i+=1      
```

```python
output= pd.DataFrame(output)
```
